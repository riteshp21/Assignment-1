```python
def outlier_remove(df1,col):
    IQR = df1[col].quantile(0.75) - df1[col].quantile(0.25)
    lower_limit = df1[col].quantile(0.25) - (IQR*1.5)
    upper_limit = df1[col].quantile(0.75) + (IQR*1.5)
    return lower_limit, upper_limit
```

```python
outliers_columns = ['Temperature', 'Ws', 'Rain', 'FFMC','DMC','ISI','BUI','FWI']
```

```python
for i in outliers_columns:
    lower,upper = outlier_remove(df1,i)
    outlier = np.where(df1[i] > upper, True, np.where (df1[i] < lower, True, False))
    df_outlier = df1.loc[outlier,i]
    df_trimed = df1.loc[~outlier,i]
    df1[i] = df_trimed
    
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
#Splitting data into Tain Test
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
#Feature Scaling
```

```python
def feature_scale(X_train,X_test):
    scaler=StandardScaler()
    X_train_scaled=scaler.fit_transform(X_train)
    X_test_scaled=scaler.transform(X_test)
    return X_train_scaled, X_test_scaled
```

```python
X_train_scaled, X_test_scaled = feature_scale(X_train,X_test)
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(data=X_train_scaled)
```

```python
#Logistic Regression
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train_scaled,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy of Logistic Regression Classifier on training set{:.4f} :  ".format(log_reg.score(X_train_scaled,y_train)))
print("Accuracy of Logistic Regression Classifier on test set{:.4f} :  ".format(log_reg.score(X_test_scaled,y_test)))
```

```python
Logistic_Regression_Prediction = log_reg.predict(X_test_scaled)
```

```python
predicted_df = pd.DataFrame({"Actual":y_test,"Predicted":Logistic_Regression_Prediction})
```
